```python
data = pd.read_csv('Transformed_Housing_Data2.csv')
```

# Scaling the dataset

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
Y = data['Sale_Price']
X = scaler.fit_transform(data.drop(columns = ['Sale_Price']))
X = pd.DataFrame(data = X, columns = data.drop(columns = ['Sale_Price']).columns)
X.head()
```

```python
X = data.drop(columns = ["Sale_Price"])
Y = data['Sale_Price']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 101)

x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# Import necessary libraries
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression

# Define a function to calculate adjusted R-squared
def adj_r2(m, data_y, predictions):
    # Calculate R-squared
    R = r2_score(data_y, predictions)
    n = len(data_y)
    # Calculate adjusted R-squared
    return 1 - ((1 - R) * (n - 1) / (n - m - 1))

# Define a function to return adjusted R-squared corresponding to a model
def return_r2(X, Y, model):
    # Fit the model
    model.fit(X, Y)
    # Make predictions
    pred = model.predict(X)
    # Calculate adjusted R-squared
    return adj_r2(len(X.columns), Y, pred)

# Define a function for forward selection
def forward_selection(model, X, Y, max_features):
    # Initialize lists to store selected features and history of Rmax
    f_col = []
    history = []

    # Iterate over the maximum number of features
    for i in range(max_features):
        # Add a placeholder for the new feature
        f_col.append('blank')

        # Create a temporary list of features
        tmp = f_col[:]

        # Initialize the maximum adjusted R2 value
        Rmax = -1

        # Iterate over each feature in the dataset
        for var in X.columns:
            # Update the temporary list with the current feature
            tmp[-1] = var

            # Calculate adjusted R2 using the temporary list of features
            r2 = return_r2(X[tmp], Y, model)

            # Check if the new adjusted R2 is greater than the maximum
            if r2 >= Rmax:
                Rmax = r2
                # Update the selected feature
                f_col[-1] = var
                # Record the feature to eliminate
                elim = var

        # Print the selected feature and its corresponding adjusted R2
        print("Selected", f_col[-1], "with new r2 =", Rmax)

        # Append the maximum adjusted R2 to the history list
        history.append(Rmax)

    # Return the final adjusted R2, history, and the dataset with the selected features
    return Rmax, history, X[f_col]

# Create a Linear Regression model instance
lr = LinearRegression()

# Perform forward selection
score, history, selected_data = forward_selection(lr, x_train, y_train, 25)
```

```python
# Import necessary libraries
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression

# Define a function to calculate adjusted R-squared
def adj_r2(m, data_y, predictions):
    # Calculate R-squared
    R = r2_score(data_y, predictions)
    n = len(data_y)
    # Calculate adjusted R-squared
    return 1 - ((1 - R) * (n - 1) / (n - m - 1))

# Define a function to return adjusted R-squared corresponding to a model
def return_r2(X, Y, model):
    # Fit the model
    model.fit(X, Y)
    # Make predictions
    pred = model.predict(X)
    # Calculate adjusted R-squared
    return adj_r2(len(X.columns), Y, pred)

# Define a function for backward elimination
def backward_elimination(model, X, Y, min_features):
    # Make a copy of the independent variables
    elim_data = X.copy()
    # Initialize a list to store the history of adjusted R-squared values
    r2_history = []
    # Calculate adjusted R-squared using all features
    base_r2 = return_r2(elim_data, Y, model)
    # Print the initial adjusted R-squared
    print("base_r2 is", base_r2)

    # Iterate until the desired minimum number of features is reached
    for i in range(X.shape[1] - min_features):
        # Initialize a variable to keep track of the maximum increase in adjusted R-squared
        difference = -100
        # Iterate over each feature
        for var in elim_data.columns:
            # Create a temporary dataset by removing the current feature
            tmp = elim_data.drop(columns=[var])
            # Calculate adjusted R-squared using the temporary dataset
            r2 = return_r2(tmp, Y, model)
            # Check if removing this feature results in a greater increase in adjusted R-squared
            if (r2 - base_r2) > difference:
                difference = r2 - base_r2
                eliminate = var
                new_r2 = r2
        # Remove the feature that results in the maximum increase in adjusted R-squared
        elim_data.drop(columns=[eliminate], inplace=True)
        # Append the new adjusted R-squared to the history list
        r2_history.append(new_r2)
        # Print the eliminated feature and the new adjusted R-squared
        print(eliminate, "Eliminated: New adj_r2 =", new_r2)
        # Update the base adjusted R-squared for the next iteration
        base_r2 = new_r2

    # Return the final adjusted R-squared, history, and the dataset with the remaining features
    return base_r2, r2_history, elim_data

# Create a Linear Regression model instance
lr = LinearRegression()
# Perform backward elimination
score, history, survived_data = backward_elimination(lr, x_train, y_train, 15)
```
